```python
#PHSX 216 Projectile Motion Lab plotting code
#performs 2nd degree polynomial fit
#Brianna Beller
#Updated September 2024

import numpy as np
import matplotlib.pyplot as plt

#-----------------------------------------------------------------------#
#----------UPDATE THIS SECTION!!!----------

#DATA TO BE PLOTTED
#UPDATE THE VARIABLE NAMES & NUMBERS (LEAVE BRACKETS/PARENTHESES AS IS)
#variable names should start with a letter and must contain only letters, numbers, and underscores

degrees = np.array([38, 41, 44, 47, 50]) #what are units?   <-----

avg_distance = np.array([0.8592,	0.9008,	0.8832,	0.7884,	0.6358]) #meters   <-----

err_avg_distance = np.array([0.009713496,	0.009179978,	0.00855757,	0.008102345,	0.02334335])                   #meters                 <-----

#Re-assign variables as x, y, dy so that the remaining plotting code may remain generic
#UPDATE VARIABLE NAMES (ON RHS OF EACH EQUAL SIGN) TO MATCH VARIABLE NAMES FOR DATA ABOVE

x = degrees             #this should be the array you want to plot on the x axis #        <-----
y = avg_distance             #this should be the array you want to plot on the y axis #        <-----
dy = err_avg_distance        #this should be your error in y array                    #        <-----

#UPDATE EACH OF THESE 3 STRINGS FOR YOUR PLOT LABELS AS YOU WANT THEM TO APPEAR

plot_title= 'Projectile Motion'                                                                #                 <-----
x_label= 'Angle [degrees]'                                                   #                 <-----
y_label= 'Average Distance [meters]'                                                    #                 <-----

#-----------------------------------------------------------------------#
#-----DON'T NEED TO CHANGE ANYTHING BEYOND THIS POINT!!!-----


#size the plot
plt.figure(figsize=(15,10))

#create scatter plot
plt.scatter(x, y, color='blue', marker='o')

#create labels
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.title(plot_title)

#fitting to a 2nd degree polynomial
c,b,a=np.polynomial.polynomial.polyfit(x,y,2,w=dy)

#Annotate with values of A, B, C from best fit polynomial
plt.annotate('best fit line: $f(x)=Ax^2+Bx+C$',(0.05, 0.45), xycoords='axes fraction')

plt.annotate('     A = {:.6f}'.format(a),(0.05, 0.4), xycoords='axes fraction')

plt.annotate('     B = {:.6f}'.format(b),(0.05, 0.35), xycoords='axes fraction')

plt.annotate('     C = {:.6f}'.format(c),(0.05, 0.3), xycoords='axes fraction')

#Create fit line
xnew = np.linspace(x.min(), x.max(), 300)
fit = a*xnew**2 + b*xnew +c

plt.plot(xnew, fit, color='green')
plt.show()
```
